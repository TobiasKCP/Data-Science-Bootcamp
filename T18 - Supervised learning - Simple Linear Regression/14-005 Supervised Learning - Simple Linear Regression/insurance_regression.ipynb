```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
our_data = pd.read_csv("insurance.csv")
our_data.head()
```

```python
  
x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
charges_model = LinearRegression()
charges_model.fit(x,y)
```

```python
unk_x = [[90]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)
print("The charges should be:", charges_model.predict(unk_x))
```

```python

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
